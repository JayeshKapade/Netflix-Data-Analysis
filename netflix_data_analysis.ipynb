# 🎬 Netflix Data Analysis (EDA)

##  Objective

To analyze Netflix dataset and understand content trends, types, and growth over time.


#IMPORT LIBRARIES

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#LOAD DATA

```python
df = pd.read_csv("netflix_titles.csv")

df.head(5)
```

#Understanding the Dataset

##We check the structure, columns, and missing values.

```python
df.info()
df.isnull().sum()
```

#Data Cleaning

##Handling missing values and removing duplicates.

```python
df['director'] = df['director'].fillna('Unknown')
df['cast'] = df['cast'].fillna('Unknown')
df['country'] = df['country'].fillna('Unknown')

df = df.dropna(subset=['date_added'])

df = df.dropna(subset=['rating', 'duration'])
```

#Date Conversion

##Convert date column to datetime for trend analysis.


```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')

df = df.dropna(subset=['date_added'])

df['year_added'] = df['date_added'].dt.year
```

#Data Analysis

##Analyzing content distribution and trends.

##Movies vs TV Shows

```python
df['type'].value_counts()
```

##Top Countries

```python
df['country'].value_counts().head(10)
```

##Year-wise Growth

```python
df['year_added'].value_counts().sort_index()
```

#Visualization

##Creating graphs to better understand data.

##Movies vs TV Shows

```python
df['type'].value_counts().plot(kind='bar')
plt.title("Movies vs TV Shows")
plt.show()
```

##Top Countries

```python
df['country'].value_counts().head(10).plot(kind='barh')
plt.title("Top Countries")
plt.show()
```

##Year-wise Growth

```python
df['year_added'].value_counts().sort_index().plot(kind='line')
plt.title("Content Over Years")
plt.show()
```

#INSIGHTS

💠Netflix has more Movies than TV Shows

💠Most content comes from a few countries

💠Content increased over time

💠Growth is visible in recent years
